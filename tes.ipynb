```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f (x):
    return (x**4) - (6.4*x**3) + (6.45**2) + (20.538*x) - 31.752
```

```python
x = np.arange (-10, 10, 0.01)
y = f(x)

plt.grid ()
plt.ylim (-100, 100)
plt.xlim (-50, 50)
plt.plot (x, y)
plt.show ()
```
